```python
def convertRange(x):
    temp=x.split('-')
    if len(temp)==2:
        return (float(temp[0])+float(temp[1]))//2
    try:
        return float(x)
    except:
        return None
        
```

```python
df['total_sqft']=df['total_sqft'].apply(convertRange)
```

```python
def remove_outliers_sqft(df):
    df_output=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_output=pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
df=remove_outliers_sqft(df)
df.describe()
        
    
```

```python
df.shape
```

```python
def bhk_remove_outlier(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.mean(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
                
                

    
```

```python
bhk_remove_outlier(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
df.to_csv("data1.csv")
X=df.drop(columns=["price"])
y=df["price"]
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


column_trans = make_column_transformer(
    (OneHotEncoder(sparse=False), ['location']),
    remainder='passthrough'
)
```

```python
lasso=Lasso()
pipe = make_pipeline(column_trans, StandardScaler(), lasso)

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Predict on the test set
y_pred = pipe.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("r2 Error:", r2)
```

```python
ridge=Ridge()
pipe = make_pipeline(column_trans, StandardScaler(), ridge)

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Predict on the test set
y_pred = pipe.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("r2 Error:", r2)
```

```python
# Split the dataset into train and test sets


linear_regression = LinearRegression()
decision_tree = DecisionTreeRegressor()
random_forest = RandomForestRegressor()

# Create an ensemble regressor
ensemble_regressor = VotingRegressor(estimators=[
    ('lr', linear_regression),
    ('dt', decision_tree),
    ('rf', random_forest)
])

pipe = make_pipeline(column_trans, StandardScaler(), ensemble_regressor)

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Predict on the test set
y_pred = pipe.predict(X_test)


r2 = r2_score(y_test, y_pred)
print("r2 Error:", r2)
```
